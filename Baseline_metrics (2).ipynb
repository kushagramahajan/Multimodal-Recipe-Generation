```python
ig=[]
tl=[]
vb=[]
for i in range (0,13551):
# for i in range (0,2):
    a=[]
    c=set()
    for j in range(0,len(based[i]['reference']['instructions'])):
        a.append(based[i]['reference']['instructions'][j]['text'])
    s_a = re.sub(r'[^\w\s]',' ',('').join(a)).lower()
    s_a=s_a.split()
    c.update(s_a)
    b=[]
    d=set()
    for j in range(0,len(based[i]['generation']['result']['recipe'])):
        b.append(based[i]['generation']['result']['recipe'][j])
    s_b = re.sub(r'[^\w\s]',' ',('').join(b)).lower()
    s_b=s_b.split()
    d.update(s_b)
    e=c.intersection(tool)
    f=d.intersection(tool)
    g=c.intersection(verb)
    h=d.intersection(verb)
    l=c.intersection(ing_set)
    m=d.intersection(ing_set)
    # print(e,f,g,h,l,m)
    if(len(e.union(f))==0):
        tl.append(0)
    else:
        tl.append(len(e.intersection(f))/len(e.union(f)))
    if(len(g.union(h))==0):
        vb.append(0)
    else:
        vb.append(len(g.intersection(h))/len(g.union(h)))
    if(len(l.union(m))==0):
        ig.append(0)
    else:
        ig.append(len(l.intersection(m))/len(l.union(m)))
```

```python
bl=[]
bl_1=[]
bl_2=[]
bl_3=[]
bl_4=[]
rg_1=[]
rg_2=[]
rg_l=[]
rg_lsum=[]
mt=[]
for i in range (0,13551):
    a=[]
    b=[]
    for j in range(0,len(based[i]['reference']['instructions'])):
        a.append(based[i]['reference']['instructions'][j]['text'])
    s_a = (" ").join(re.sub(r'[^\w\s]',' ',('').join(a)).lower().split())
    for j in range(0,len(based[i]['generation']['result']['recipe'])):
        b.append(based[i]['generation']['result']['recipe'][j])
    s_b = (" ").join(re.sub(r'[^\w\s]',' ',('').join(b)).lower().split())
    bl_r=bleu.compute(predictions=[s_b], references=[s_a])
    bl.append(bl_r['bleu'])
    bl_1.append(bl_r['precisions'][0])
    bl_2.append(bl_r['precisions'][1])
    bl_3.append(bl_r['precisions'][2])
    bl_4.append(bl_r['precisions'][3])
    rg_r=rouge.compute(predictions=[s_b], references=[s_a])
    rg_1.append(rg_r['rouge1'])
    rg_2.append(rg_r['rouge2'])
    rg_l.append(rg_r['rougeL'])
    rg_lsum.append(rg_r['rougeLsum'])
    mt_r=meteor.compute(predictions=[s_b], references=[s_a])
    mt.append(mt_r['meteor'])

```
